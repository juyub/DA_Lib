## 1-1. 벡터

* 스칼라 (scalar)

      - V, +, · 의 세 요소가 정의되는 성립되는 공간

      - +: 덧셈 규칙
      - ·: 곱셈 규칙
    
* 벡터의 표현
    - 파이썬 리스트
    - 넘파이 배열

## 1-3. 벡터의 선형조합 (스칼라와 벡터의 곱)

#### 벡터의 상수배


### 1-3-1. sum 함수, zip 함수 이용

```python
Hong1=[80, 90]
Jung1=[70, 80]
a=0.5

aa = [a*sum(i) for i in zip(Hong1,Jung1)]
aa
```

```python
aa_lam = list(map(lambda x : x*0.5, [sum(i) for i in zip(Hong1, Jung1)]))
aa_lam
```

### 1-3-2. map, lambda 함수 사용
* **lambda** 함수 : **lambda** *매개변수* **:** *표현식*

```python
# map(함수, data)
# lambda x : x*0.5
# map(lambda x : x*0.5, data)
```

```python
A=[1, 2, 3, 4, 5]
aaa = list(map(lambda x : x*0.5, A))
aaa
```

## 1-6. 벡터를 이용한 평균 제곱 오차와 평균 제곱근 오차
### 1-6-1. 평균 제곱 오차

```python
y=[1, 1, 0, 1, 0, 1, 0, 1, 0, 0] # 실제값
t=[0, 0, 1, 0, 0, 1, 0, 1, 0, 0] # 예측값
y_np = np.array(y)
t_np = np.array(t)
mse = np.sum((y_np-t_np)**2)/len(y)
mse
```

### 1-6-2. 평균 제곱근 오차

```python
c=np.array([1, 2, 3, 4, 5])
d=np.array([2, 4, 6, 8, 10])
msre = np.sqrt(sum((c-d)**2)/len(c))
msre
```
